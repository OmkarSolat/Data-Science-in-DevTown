# C-> CREATE

```python
data = {"Apples":[102, 500, 250, 75],"Grapes":[50, 150, 90, 45]}
df = pd.DataFrame(data,index=["A", "B", "C", "D"])
df
```

```python
df['Apples']           #Series
```

```python
df[['Apples']]           #DtaFrame
```

```python
df['Bananas'] = df['Grapes'] + 50
df
```

```python
df.drop('Bananas',axis=1)      #temporary
```

```python
df
```

```python
df.drop('Bananas',axis=1,inplace=True)    # Parmanent
```

```python
df
```

```python
df['Pear'] = df['Apples'] + 100
df['Mangoes'] = df['Pear'] + 500
df['Jackfruit'] = df['Grapes'] + 150
df['Strawberry'] = df['Jackfruit'] + 200
df['Oranges'] = df['Strawberry'] + 70
```

```python
df
```

# U -> UPDATE

* Row addition

```python
student_marks = {"Name":["Siraj", "Virat", "Ram", "Sham", "Sita"],
                 "OS":[25, 75, 96, 65, 100],
                 "IOT":[20, 80, 60, 50, 99]
                }

df1 = pd.DataFrame(student_marks)
```

```python
df1
```

```python
data2 = {"Name":"Root","OS":81,"IOT":96}
```

```python
df1.append(data2,ignore_index=True)
```

```python
data3 = {"Name":"Rohit","OS":48,"IOT":45}
df2 = pd.DataFrame(data3,index = [5])
```

```python
df2
```

```python
df1 = pd.concat([df1,df2])           #concatenation of two DataFrame
```

```python
df1
```

```python
df1.loc[[5]]
```

```python
df1.iloc[[5]]
```

```python
df1
```

```python
df1.loc[5] = ["Smith",49,49]
```

```python
df1
```

```python
df1.iloc[5] = ["Dhoni",77,77]
```

```python
df1
```
