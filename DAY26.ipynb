```python
#Reference of lines
'''
- -> Solid Line
: -> Dotted Line
-- -> Dashed Line
-. -> Dashed/Dotted Line

'''
```

* Format String (fmt)

```python
# SYNTAX: marker|line|color
```

```python
# Reference of Colors
'''
r -> Red
g -> Green
b -> Blue
y -> Yellow
w -> white
k -> Black

'''
```

```python
plt.plot(x,y,'o--r')
```

```python
plt.plot(x,y,marker='*',ms = 30,mec='r',mfc='y')
 # ms -> marker size
 # mec -> marker edge color
 # mfc -> marker face color
```

* Labels

```python
plt.plot(x)
plt.plot(y)

plt.xlabel('Height')
plt.ylabel('Weight')

plt.show()
```

* Title

```python
plt.plot(x)
plt.plot(y)

modification = {"color":"red",'size':15,"family":"serif"}

plt.title(" Height V/S Weight Graph",loc='left',fontdict=modification)

plt.xlabel('Height',fontdict=modification)
plt.ylabel('Weight',fontdict=modification)

plt.show()
```

* Grid

```python
plt.plot(y)
plt.plot(x)

plt.xlabel("Height")
plt.ylabel("Weight")

plt.grid()

plt.show()
```

```python
plt.plot(y)
plt.plot(x)

plt.xlabel("Height")
plt.ylabel("Weight")

plt.grid(color="black",linewidth="1.5",ls=":",axis="x")

plt.show()
```

* Ticks  & Ticks Label

```python
x = [5,12,9,18,7,12]
y = [5,20,14,4,17,9]

fig = plt.figure()
ax = fig.add_axes([1,1,1.2,1.2])             #add_axes([x1,x2,y1,y2])

ax.plot(x,y)
ax.set_xlabel("X-axis",color="green")
ax.set_ylabel("Y-axis",color="green")
ax.set_title("MAVERICKS",color="red")

ax.set_xticklabels(['A','B','C','D','E','F'])
ax.set_yticklabels([101,102,103,104,105,106])
```

# Pie chart

```python
x= [60,20,15,5]

plt.pie(x,
        labels=["KHO-KHO","KABADDI","CRICKET","FOOTBALL"],
        colors=["orange","red","blue","brown"],
        shadow=True,
        explode=[0,0,0,0]
        )

plt.legend(title="GAMES OF SPIRIT")
```
