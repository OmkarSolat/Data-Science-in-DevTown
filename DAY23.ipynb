# C -> Creation

* ways to create DataFrame:

1. List
2. Dict

4. Series
5. Another DataFrame

* Empty DataFrame

```python
df = pd.DataFrame()
df
```

* Create a DataFrame using List

```python
data = [1,4,6,8,2,3,0,8]
df = pd.DataFrame(data)              #Dataframe
df
```

```python
s = pd.Series(data)                #Series
s
```

```python
data = [['prachi','sanchita','tanishka'],['sai','shruti','divya'],['shravani','gauri','swati']]
df = pd.DataFrame(data)
df
```

```python
data = [['prachi','sanchita','tanishka'],['sai','shruti','divya'],['shravani','gauri','swati']]
df = pd.DataFrame(data,index=['batch1','batch2','batch3'],columns=['c1','c2','c3'])
df
```

* Create a DataFrame using Dictionary(Dict)

```python
data = {"Name":["Ram", "sham", "Sita"], "Age":[25, 20, 18]}
df = pd.DataFrame(data)
df
```

```python
data = {"Name":["Ram", "sham", "Sita"], "Age":[25, 20, 18]}
df = pd.DataFrame(data,index=['A','B','C'])
df
```

* Create DataFrame using Series

```python
data = {"OS_marks":pd.Series([90,92,96,85,89],index=['Test1','Test2','Test3','Test4','Test5']),
        "OD_marks":pd.Series([96,93,84,65,94],index=['Test-1','PTest1','Test2','PTest2','Test3'])}

df = pd.DataFrame(data)
df


# NaN ----> Not a Number
```

```python
data = {"Karan_Marks":pd.Series([50,77,65,98],index=["Test1","Test2","Test3","Test4"]),
        "Arjun_Marks":pd.Series([67,89,32,98,50],index=["Test-1","PTest1","Test-2","PTest2","Test-3"])}
df = pd.DataFrame(data)
df
```

* Create a DataFrame from List of Dict

```python
data = [{'A':18,'B':45,'P':100},{'C':7,'D':17},{'E':33,'F':10}]
#          index=0         index=1          index=2
df = pd.DataFrame(data)
df
```

```python
data = [{'A':18,'B':45,'P':100},{'C':7,'D':17},{'E':33,'F':10}]
df = pd.DataFrame(data,index=['X','Y','Z'],columns=['p','q','r','s'])
df
```
